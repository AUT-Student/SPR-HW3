# Libraries

```python
import pandas as pd
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

# Load Dataset

```python
data = pd.read_csv("./BOM.csv")
```

```python
data
```

# Preprocessing

```python
data = data.dropna(axis=0, subset=["RainTomorrow"])
```

```python
data["Month"] = data["Date"].apply(lambda x:int(x[5:7]))
```

```python
data[['MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation', 'Sunshine', 'WindGustSpeed', 'WindSpeed9am',
      'WindSpeed3pm', 'Humidity9am', 'Humidity3pm','Pressure9am', 'Pressure3pm', 'Cloud9am', 'Cloud3pm', 'Temp9am',
       'Temp3pm']] = data.groupby("Month").transform(lambda x:x.fillna(x.median()))
```

```python
data[['WindGustDir', 'WindDir9am', 'WindDir3pm'
      ,'RainToday']] = data.groupby("Month").transform(lambda x:x.fillna(x.mode().iloc[0]))[['WindGustDir', 'WindDir9am', 'WindDir3pm'
      ,'RainToday']]

# data.fillna(data.mode().iloc[0])
```

```python
data.isna().any()
```

```python
data[['WindGustDir', 'WindDir9am', 'WindDir3pm','RainToday','RainTomorrow']]
```

```python
data =  data.drop(columns=["Date", "Location"])
```

```python
data["RainToday"] = data["RainToday"].map({"No":0, "Yes":1})
data["RainTomorrow"] = data["RainTomorrow"].map({"No":0, "Yes":1})

direction_dict = {'E':0, 'ENE':1, 'ESE':2, 'N':3,
                  'NE':4, 'NNE':5, 'NNW': 6, 'NW':7,
                  'S':8, 'SE':9, 'SSE':10, 'SSW':11,
                  'SW':12, 'W':13, 'WNW':14,'WSW':15}

data["WindGustDir"] = data["WindGustDir"].map(direction_dict)
data["WindDir9am"] = data["WindDir9am"].map(direction_dict)
data["WindDir3pm"] = data["WindDir3pm"].map(direction_dict)
```

# Train, Test and Validation Split

```python
data = data.sample(frac=1, random_state=0)
number_of_data = len(data)

data_train = data[0: int(number_of_data*0.6)]
data_validation = data[int(number_of_data*0.6): int(number_of_data*0.8)]
data_test = data[int(number_of_data*0.8): ]
```

# a

```python
data_train.corr()
```

```python
data_train_corr = data_train.corr().abs()

print(data_train_corr.unstack().sort_values(kind="quicksort", ascending=False)[:30])
```

```python
redundant_features = ["MaxTemp", "Pressure9am"]
data_train = data_train.drop(columns=redundant_features)
data_validation = data_validation.drop(columns=redundant_features)
data_test = data_test.drop(columns=redundant_features)
```

# b

```python
sfs_knn = KNeighborsClassifier(n_neighbors=11)
sfs = SequentialFeatureSelector(sfs_knn, n_features_to_select=10)
sfs.fit(data_train.drop(columns=["RainTomorrow"]), data_train["RainTomorrow"])
```

```python
data_train.drop(columns=["RainTomorrow"]).columns[sfs.get_support()]
```

# c

```python
data_train_reduced_x = sfs.transform(data_train.drop(columns=["RainTomorrow"]))
data_train_reduced_y = data_train["RainTomorrow"]

data_test_reduced_x = sfs.transform(data_test.drop(columns=["RainTomorrow"]))
data_test_reduced_y = data_test["RainTomorrow"]

data_validation_x = sfs.transform(data_validation.drop(columns=["RainTomorrow"]))
data_validation_y = data_validation["RainTomorrow"]
```

```python
k_list = []
accuracy_list = []

best_accuracy = 0
best_k = None

for k in range(1, 41, 2):

    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(data_train_reduced_x, data_train_reduced_y)

    y_real = data_validation_y
    y_predict = knn.predict(data_validation_x)

    score = accuracy_score(y_real, y_predict)

    if score>best_accuracy:
        best_accuracy = accuracy
        best_k = k
    
    k_list.append(k)
    accuracy_list.append(score)
    
print(f"Best k = {k} with accuracy {best_accuracy}")
```

```python
plt.scatter(k_list, accuracy_list)

plt.xlabel("K")
plt.ylabel("Accuracy Score")
plt.title("Accuracy Score for Validation Set Over different K")
plt.show()
```

```python
knn = KNeighborsClassifier(n_neighbors=best_k)
knn.fit(data_train_reduced_x, data_train_reduced_y)

y_real = data_test_y
y_predict = knn.predict(data_test_x)

score = accuracy_score(y_real, y_predict)

print(f"Accuracy for Test Set = {score}")
```
